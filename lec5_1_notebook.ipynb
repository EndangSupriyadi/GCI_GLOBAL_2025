### 4-1.2.2 Adjusting Graphs

In the next code cell, we'll explore how to customize various elements of a graph using Matplotlib.

**Graph Title and Axis Labels**

*   `plt.title()`: Adds a title displayed at the top of the graph.
*   `plt.xlabel()`: Sets the label for the X-axis.
*   `plt.ylabel()`: Sets the label for the Y-axis.


These functions take the text you want to display as a string argument.

**Grid Lines**

*   `plt.grid(True)`: Displays grid lines in the background of the graph.

You can toggle grid lines on and off by passing True/False as an argument.

**Graph Size Adjustment**

*   `plt.figure(figsize=(width, height))`: Customizes the width and height of the graph.
Arguments specify width and height in order, typically in inches.

```python
# Load data
iris = sns.load_dataset('iris')

# Set x/y-axis
x = iris['sepal_length']
y = iris['petal_width']

# Specify the size of the graph (try changing 10 and 6）
plt.figure(figsize=(10, 6))

# Plot the graph
plt.scatter(x, y)

# Title
plt.title('Sepal length vs. Petal width')
# Label for the x-axis
plt.xlabel('sepal_length')
# Label for the y-axis
plt.ylabel('petal_width')

# Display grid lines in the graph
plt.grid(True)
```


**Font Size Customization**

Many functions (title(), xlabel(), ylabel(), etc.) have a fontsize argument that allows you to adjust the text size.

When making the graph larger, proportionally increase the font size for better readability.
Refer to the official Matplotlib documentation for more detailed customization options.

```python
# Load data
flight = sns.load_dataset('flights')

# Set x/y-axis
x = np.arange(flight.shape[0])
y = flight['passengers']

# Specify the size of the graph
plt.figure(figsize=(15, 4.5))

# Add a label using label= and create a legend
plt.plot(x, y, label='Number of passengers')
plt.legend(fontsize=20)

# Specify labels and change the font size
plt.title('Number of passengers by elapsed months', fontsize=20)
plt.xlabel('Elapsed months since Jan 1949', fontsize=20)
plt.ylabel('Number of passengers', fontsize=20)

# Change the font size of the ticks
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.grid(True)
```

### 4-1.2.3 Graph Partitioning

Using `plt.subplot()`, you can divide a graph into multiple sections. The following example demonstrates how to create a graph with 2 rows and 1 column, displaying the first and second plots based on the specified numbering. Note that `np.linspace(-10,10,100)`generates a list of 100 evenly spaced numbers ranging from -10 to 10.

```python
# Specify the size of the graph
plt.figure(figsize=(20, 6))

# First graph in a 2×1 grid
plt.subplot(2,1,1)
x = np.linspace(-10, 10, 100)
plt.plot(x, np.sin(x))

# Second graph in a 2×1 grid
plt.subplot(2,1,2)
y = np.linspace(-10, 10, 1000)
plt.plot(y, np.sin(2*y))

plt.grid(True)
```

### 4-1.2.6 Histogram

The following graph is called a **histogram**, which displays the **frequency** (the nubmer of occurrences) of each value. It is commonly used when observing the overall distribution of data. By using `plt.hist()` as shown below, you can create a histogram. The frequently-used arguments are `bins` and `range`. You can set `bins` to specify the number of bins or their edges. You can set `range` to specify the lower and upper range of the bins.

### 4-1.2.7 Bar Charts

Next, we will discuss **bar charts**, which are used to compare values across different categories (such as by region or department). To display a bar chart, you can use the `plt.bar()` function. If you wish to display labels on the bars, you can specify them using the `xticks` functoin as shown below.

Additionally, when executed without adjustments, the graph will be left-aligned. To center the graph, it is advisable to use the `align = 'center'` argument.

```python
# Data to be displayed
x = [1, 2, 3]
y = [10, 1, 4]

# Specify the size of the graph
plt.figure(figsize = (10, 6))

plt.bar(x, y, align='center', width=0.5)

# Labels for each bar in the bar chart
plt.xticks(x, ['A Class', 'B Class', 'C Class'])

# Set the labels for x and y axes
plt.xlabel('Class')
plt.ylabel('Score')

# Display the grid
plt.grid(True)
```

#### Horizontal Bar Chart
While the previous example used a vertical bar chart, you can create a horizontal bar chart by using the `plt.barh()` function. Note that the x and y axes will switch places, so you'll need to reset the labels accordingly.

```python
# Data to be displayed
x = [1, 2, 3]
y = [10, 1, 4]

# Specify the size of the grpah
plt.figure(figsize = (10, 6))

plt.barh(x, y, align='center')
plt.yticks(x, ['A Class','B Class','C Class'])
plt.ylabel('Class')
plt.xlabel('Score')
plt.grid(True)
```

#### Drawing Multiple Graphs

Now, we will create multiple bar graphs to use for comparison. The example below illustrates the first-term and final-term grades for each class, providing an easy way to compare the results.

```python
# Preparing the data
y1 = np.array([30, 10, 40])
y2 = np.array([10, 50, 90])

# Data for the x-axis
x = np.arange(len(y1))

# Width of the bars
w = 0.4

# Specify the size of the graph
plt.figure(figsize = (10, 6))

# Drawing the graphs. The second set (y2) is shifted to the right by the width of the bars
plt.bar(x, y1, color='blue', width=w, label='First', align='center')
plt.bar(x + w, y2, color='green', width=w, label='Final', align='center')

# Set title
plt.title("Comparison between first and final math grades")

# Place the legend in the best position
plt.legend(loc='best')

plt.xticks(x + w / 2, ['Class A', 'Class B', 'Class C'])
plt.grid(True)
```

#### Stacked Bar Graph

The following is an example of a stacked bar graph. Although we are also using the `plt.bar()` function here, pay attention to the setting of the `bottom` argument. For the graph being stacked on top, specify `bottom=<the graph below>` as an argument in the `plt.bar()` function.

```python
# Prepare the data
height1 = np.array([100, 200, 300, 400, 500])
height2 = np.array([1000, 800, 600, 400, 200])

# X-axis
x = np.array([1, 2, 3, 4, 5])

# Specify the size of the graph
plt.figure(figsize = (10, 6))

# Draw the graph
p1 = plt.bar(x, height1, color='blue', label="Class 1")
p2 = plt.bar(x, height2, bottom=height1, color='lightblue', label="Class 2")

# Display the legend
plt.legend(loc='best')
```

### 4-1.2.8 Pie Charts and Bubble Charts
Next, we will explain how to create **pie charts** and **bubble charts**. Pie chart is used to see how each component contributes to the whole in terms of percentage, while bubble chart is a variant of the scatter plot that adds size information for each data.

#### Pie Charts
To draw a pie chart, use the `plt.pie()` function. You can customize by setting labels, colors, and other arguments. For example, the `autopct` argument allows you to specify the format for displaying each percentage. Also, you can use the `explode` argument to separate a specific category from the overall pie chart (in this case, `Hogs` is set to 0.1.)

The `startangle` argument represents the angle at which the output of each element begins. By specifying this parameter, you can change the starting position of the output. For example, setting it to "`90`" makes the starting point at the top. To change the direction to counterclockwise, specify a positive value; to make it clockwise, specify a negative value.

The direction of the output can also be specified with the `counterclock` argument. If set to `True` or left unspecified, the output will be counterclockwise. If set to `False`, it will be displayed clockwise.

The `plt.axis()` function can be used to adjust the chart for a circular display.

```python
import matplotlib.pyplot as plt

labels = ['Frogs', 'Hogs', 'Dogs', 'Logs']
sizes = [15, 30, 45, 10]
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
explode = (0, 0.1, 0, 0)

# Specify the size of the chart
plt.figure(figsize = (15, 6))

# Display the pie chart
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)

# Draw the pie chart as a circle
plt.axis('equal')
```

You can also check the parameters of the `pie` function using the `help` function.

```python
help(plt.pie)
```

#### Bubble Chart
Next, let's create a **bubble chart** using the `plt.scatter()` function. A bubble chart is a graph that uses bubbles to display three dimensions of data. It's a variation of a scatter plot, where the data points are replaced with bubbles.

```python
N = 25

# Generate random X and Y data
x = np.random.rand(N)
y = np.random.rand(N)

# Color numbers
colors = np.random.rand(N)

# Vary the size of the bubbles
area = 10 * np.pi * (15 * np.random.rand(N)) ** 2

# Specify the size of the grpah
plt.figure(figsize = (15, 6))

# Draw the graph
plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.grid(True)
```

You can also check the parameters of the `scatter` function using the `help` function.

```python
help(plt.scatter)
```

## 4-1.3 Visualization Using Seaborn

**Seaborn** is a visualization library based on Matplotlib that not only enables more attractive plotting but also offers a variety of APIs that enhance the convenience of Matplotlib. You can simply replace the line of code to plot a certain type of graph in matplotlib with the seaborn's version of plottng the graph. Customizations of the graphs can be done using matplotlib's functions, ensuring a high level of compatibility. By learning both libraries together, you can effectively expand your graphing capabilities.

We usually import seaborn using the abbreviation, `sns`.  


### 4-1.3.1 Different Types of Graphs Using Seaborn

For example, we can replace `plt.scatter()` function  with seaborn's `sns.scatterplot()` function.

```python
iris = sns.load_dataset('iris')
x = iris['sepal_length']
y = iris['petal_width']

# Plotting the graph
# instead of plt.scatter(x, y), we are using sns.scatterplot(x, y)
sns.scatterplot(x=x, y=y)

# Customization can be done using matplotlib
plt.title('Relationship between Sepal Length and Petal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')
```

One advantage of seaborn is the `hue` argument. By specifying a categorical variable column that you want to use for color differentiation, you can easily compare the distributions of different categories, as shown below.   

```python
sns.scatterplot(x=x, y=y, hue=iris['species'])

plt.title('Relationship between Sepal Length and Petal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')
```

The `hue` argument looks at the specified column of the dataset and groups the dataset based on its categories.

```python
iris
```

Another advantage of seaborn is that it can directly refer to a pandas DataFrame using the `data` argument. By passing the DataFrame as `data` argument, you only need to specify the column names for `x`, `y`, and `hue`.

```python
sns.scatterplot(data=iris, x='sepal_length', y='petal_width', hue='species')

plt.title('Relationship between Sepal Length and Petal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')
```

Other graphs are also available in seaborn as well.

For example, histograms can be plotted using `sns.histplot()`. Again, we can use `hue` and `data` arguments.

```python
# Specify the size of the graph
plt.figure(figsize = (12, 8))

plt.subplot(2,1,1)
plt.hist(x=iris['sepal_length'], bins=25, range=(4.5, 7.5))
plt.xlabel('Sepal Length')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

plt.subplot(2,1,2)
sns.histplot(data=iris, x='sepal_length', hue='species', bins=25, binrange=(4.5, 7.5))
plt.xlabel('Sepal Length')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)

# Adjust the padding between and around subplots
plt.tight_layout()
```

We can also plot the **kernel density estimate (KDE)** by setting `kde=True`, which shows where the peaks of the distribution are concentrated.

```python
plt.figure(figsize = (12, 4))

sns.histplot(data=iris, x='sepal_length', hue='species', bins=25, binrange=(4.5, 7.5), kde=True)
plt.xlabel('Sepal Length')
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
```

Seaborn library also has a separate plotting function for visualizing KDE, `sns.kdeplot()`.

```python
# kdeplot
sns.kdeplot(data=iris, x='sepal_length', y='petal_width', hue='species')

plt.title('Relationship between Sepal Length and Petal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')
```

```python
# Setting date data. Generate data every minute with freq=`T`.
idx = pd.date_range('2015/01/01', '2015/12/31 23:59', freq='T')

# Generate random values, producing either 1 or -1.
dn = np.random.randint(2, size = len(idx)) * 2 - 1

# Create a random walk (data where values increase or decrease randomly)
# np.cumprod calculates the cumulative product (the product of the first element, the second element, the third element, etc.).
rnd_walk = np.cumprod(np.exp(dn * 0.0002)) * 100

# Resample the data on business days using resample(`B`).
# Use the ohlc method to obtain four data points: "open," "high," "low," and "close."
df = pd.Series(rnd_walk, index=idx).resample('B').ohlc()
```

When plotted as is, the result appears as shown below. Here, we are using Pandas' visualization feature. Since the values are generated randomly, the graph will differ from the one presented in print.

```python
df.plot(figsize = (15,6), legend = 'best', grid = True)
```

>The moving average of time series data $(\cdots ,y_{t-1},y_t,y_{t+1}, \cdots )$ over the past n periods refers to the average of data from the past n periods, which means the following:

\begin{eqnarray}
ma_t = \sum_{s=t-n+1}^t \frac{y_s}{n}
\end{eqnarray}

>The logarithmic time series data of time ratio between the current period and the previous period (1 week) for the time series data $(\cdots ,y_{t-1},y_t,y_{t+1}, \cdots )$ is the data composed of $\log \frac{y_t} {y_{t-1}}$. When the rate of change $r_t = \frac{y_t - y_{t-1}}{y_t}$ is small, the relationship $r_t \approx \log \frac{y_t} {y_{t-1}}$ holds. This is derived from the approximation $\log (1+x) \approx x$ when $x$ is sufficiently small. The volatility of the rate of change data $(r_1,\cdots ,r_N )$ refers to its standard deviation

\begin{eqnarray}
\sqrt{\frac{1}{N}\sum_{t=1}^N \left(r_t - \frac{1}{N}\sum_{t=1}^N r_t\right)^2}
\end{eqnarray}

> and is used as an indicator of the magnitude of price fluctuations.
