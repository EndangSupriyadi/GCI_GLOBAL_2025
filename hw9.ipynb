## 1. Importing Libraries

````python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
````

## 3. Solution

Please complete the following function and submit it to Omnicampus.

````python
# Only the homework function within this cell should be submitted to Omnicampus
def homework(path_winequality_data, n):
    df = pd.read_csv(path_winequality_data, sep=';')
    X = df.values  # gunakan semua kolom, termasuk 'quality'
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    kmeans = KMeans(n_clusters=n, init="k-means++", random_state=0)
    labels = kmeans.fit_predict(X_scaled)
    my_result = np.bincount(labels, minlength=n).astype(int)
    return my_result
````

````python
# Feel free to use this code block for testing
n = 10  # Set to any positive integer

homework(path_winequality_data, n)
````
